```
import cv2
import torch
import numpy as np
from facenet_pytorch import MTCNN
```

```
import os
images = ["/content/"+img for img in os.listdir("/content/") if img.endswith(".jpg")]
```

```
images
```

```
import matplotlib.pyplot as plt
import cv2
```

```
def load_img(path):
    image = cv2.imread(path)
    image = cv2.resize(image,(224, 224))
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    return image
```

```
fig = plt.figure(figsize=(10, 10))

for i in range(6):
    plt.subplot(4, 4, i+1)
    plt.imshow(load_img(images[i]), cmap='gray')
    plt.suptitle("Faces",fontsize=20)
    plt.axis('off')
```

```
mtcnn = MTCNN()

fig = plt.figure(figsize=(10, 10))

for i in range(6):
    plt.subplot(4, 4, i+1)
    imge = cv2.imread(images[i])
    imge = cv2.resize(imge,(224, 224))
    imge = cv2.cvtColor(imge, cv2.COLOR_BGR2RGB)
    boxes, probs, landmarks = mtcnn.detect(imge, landmarks=True)
    boxes = boxes[0]
    boxes = [int(i) for i in boxes]
    im = cv2.rectangle(imge,
                  (boxes[0], boxes[1]),
                  (boxes[2], boxes[3]),
                  (0, 0, 255),
                  2)
    plt.imshow(im, cmap='gray')
    plt.suptitle("Detected Faces",fontsize=20)
    plt.axis('off')
```
